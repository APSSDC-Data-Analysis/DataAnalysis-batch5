```python
import numpy as np
x = np.linspace(0,10,10)
x
```

```python
plt.plot(x)
plt.plot(x,x**2,c= "c")
plt.plot(x,x**3,c = "m")
plt.xlim(0,15)
plt.ylim(0,800)
plt.show()
# just add label title and legend
```

```python
x = np.arange(80)
plt.plot(x, x+10,"*g")
plt.plot(x, x+20,"--y")
plt.plot(x, x+30,".c")
plt.plot(x, x+40,"-.r")
# * , < ,> ^, --,-, -.   --- ect 
plt.show()
```

```python
x = np.arange(1, 15)
y = x**2
plt.plot(x,x,linewidth = 0.9,marker = "o",linestyle = "-.")
plt.plot(x,x**2,linewidth = 0.5,marker = "d",linestyle = "--")
plt.plot(x,x*2,linewidth = 0.7,marker = "<",linestyle = "-.")
plt.legend(['y = x**2',"y = x*2" , "y = x"])
plt.grid()
plt.title("X & Y")
plt.show()
```

## Subplots

```python
plt.subplots(3,2) # ROWS,COLUMNS
plt.show()
```

```python
fig,pl = plt.subplots(2,2)
x = np.arange(1,10)
pl[0][0].plot(x,x)
pl[0][0].set_title("Linear")
pl[0][1].plot(x,x**2)
pl[0][1].set_title("Squares")
pl[1][0].plot(x,x**3)
pl[1][0].set_title("Cubes")
pl[1][1].plot(x,np.log(x))
pl[1][1].set_title("Logarithm")
plt.tight_layout()
plt.show()
```

# 3. Bar Graph
* Represented in height... bars
* applications are population, stock market.. etc

```python
x = np.arange(1,100,2)
y = np.linspace(1,50,len(x))
```

```python
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "R", "C++"]
y = [80,30,55,90]
plt.bar(x,y,width = 0.2,align = "edge")
plt.show()
```

```python
x = ["python", "c", "R", "C++"]
y = [80,30,55,90]
plt.bar(x,y,width = 0.2,align = "center")
plt.show()
```

```python
plt.bar(x,y,color = ["b","g",'m',"c"])
plt.text(1,30,"C Language")  # 0,1,2,3
plt.title("Programming Languages")
plt.show()
```

```python
help(plt.bar)
```

```python
plt.barh(x,y,color = ["b","g",'m',"c"])  # horizontal bar graph
#plt.text(1,30,"C Language")  # 0,1,2,3
plt.title("Programming Languages")
plt.show()
```

```python
x = ["python", "c", "R", "C++"]
internal = [23,32,38,23]
external = [80,78,67,98]

plt.bar(x,internal)
plt.bar(x, external, bottom = internal )
plt.title("Final Result")
plt.show()
```

```python
x = ["python", "c", "R", "C++"]
internal = [23,32,38,23]
external = [80,78,67,98]

plt.barh(x, internal, fill = False, hatch = " .")
plt.barh(x,external, left = internal, fill = False,hatch = "*" )
plt.show()
```

# 5. Pie chart

```python
year = [2020,2019,2018,2017,2016]
per = [90,89,87,67,60]
plt.pie(per, labels = year)
plt.show()
```

```python
sub = ["C","R","C++","Python","ML"]
marks = [90,89,34,56,78]
plt.pie(marks,labels = sub, colors = ["c","r","m","g","black"], explode = [0,0.1,0,0,0.5],radius =1.5)
plt.show()
```

```python
sub = ["C","R","C++","Python","ML"]
marks = [90,89,34,56,78]
plt.pie(marks,labels = sub, colors = ["c","r","m","g","black"], explode = [0,0.1,0,0,0.5],
        radius =1,autopct = "%1.2f%%"
       )
plt.legend()
plt.xlim(2,5)
plt.show()
```

```python
sub = ["C","R","C++","Python","ML"]
marks = [90,89,34,56,78]
plt.pie(marks,labels = sub, colors = ["c","r","m","g","black"], explode = [0,0.1,0,0,0.5],
        radius =1,autopct = "%1.2f%%", startangle= 45
       )
plt.legend()
plt.xlim(2,5)
plt.show()
```

# 6. Box Plot
* we can easily findout the outliers 
* we can calculate min value, max value, 25 quantile, 75 quantile and median of the given data

```python
data = np.random.randint(1,100,10)
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
a = np.random.randint(50,100,10)
b = np.random.randint(40,90,10)
c = np.random.randint(50,100,10)

plt.boxplot([a,b,c], labels = ["Box1", "Box2", "Box3"], 
            patch_artist = True,boxprops = dict(facecolor = "k",color = "y"))
plt.show()
```
