## Cleaning Data Using Pandas
### Working on Duplicate Values and Missing Values
- There are mainly 4 methods to identify missing and manupulate missing values
    - isnull(): Presence of missing values, returns a boolean data
    - notnull(): Opposite to isnull()
    - dropna(): It drops null values
    - fillna(): Fills the missing values by specified values
    
    
- ***Note: Always have a backup of original data, if you are working on missing values***
- ***Warning: Do not run program more than 1 time, if you are working on drop functions***

```python
d3 = {'Names':['Harika','Harika','swetha','sampath','sampath','vamsi'],
     'Branch':['Cse','Cse','It','ECE','ECE','Mech']}
data = pd.DataFrame(d3)
```

```python
data
```

```python
# Duplicated no.of rows and column
data[data.duplicated()].shape
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates()
```
